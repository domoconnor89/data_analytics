`````python
## Recreating the parse_dataset() function and outputting 2020...

pd.read_csv(file_path, names = ['name', 'gender', 'frequency'])

year = 1880

def parse_dataset(year):
    file_path = f'../data/names/yob{year}.txt'
    df = pd.read_csv(file_path, names = ['name', 'gender', 'frequency'])
    return df
`````

`````python
# Using your parse_dataset() function write a program that calculates the total number of babys 
# for the year 2020 and prints the number to the screen. 
# Make sure to select the correct column to sum.

year = 1950

def freq_sum(year):
    """Provides the total frequency of babies for a given year"""
    data_yr = parse_dataset(year)
    pop_total = data_yr['frequency'].sum()
    return pop_total

freq_sum(2020)
`````

`````python
# Do the same for 1950 and compare the results+

freq_sum(1950)
`````

`````python
print('The difference between the populations of 1950 and 2020 is:', (freq_sum(1950)-freq_sum(2020)))
`````

`````python
# What other interesting insights can you find about about a year using some of pandas methods?
# Hint: Try combining some methods like .sort_values(), .count(), .head() and others.


def count_sum(year):
    """Provides the count of all names for a given year"""
    data_yr = parse_dataset(year)
    cnt_total = data_yr['name'].count()
    return cnt_total

count_sum(1950)

`````

`````python
count_sum(2020)
`````

`````python
def unique_names(year):
    """Provides the number of unique names for a given year, """
    data_yr = parse_dataset(year)
    unq_total = data_yr['name'].nunique()
    return unq_total

unique_names(1950)
`````

`````python
unique_names(2020)
`````

`````python
print('total number of entries in 2020:',count_sum(2020))

print('number of names not shared across genders in 2020:',unique_names(2020))

print('number of names shared across genders in 2020:',count_sum(2020) - unique_names(2020))
`````

`````python
def gender_shared_names(year):
    """Provides the number of names used for both boys and girls of a given year, """
    data_yr = parse_dataset(year)
    cnt_total = data_yr['name'].count()
    unq_total = data_yr['name'].nunique()
    gen_shr_total = cnt_total - unq_total
    return gen_shr_total

gender_shared_names(2020)
`````

`````python
def total_males(year):
    """Provides the number of males babies for a given year, """
    data_yr = parse_dataset(year)
    male_df = data_yr[data_yr['gender'] != 'F']
    ttl_males = male_df['frequency'].sum()
    return ttl_males

total_males(2020)
`````

`````python
def total_females(year):
    """Provides the number of females babies for a given year, """
    data_yr = parse_dataset(year)
    female_df = data_yr[data_yr['gender'] != 'M']
    ttl_females = female_df['frequency'].sum()
    return ttl_females

total_females(2020)
`````

`````python
print(freq_sum(2020))

print(total_males(2020))

print(total_females(2020))

print(total_males(2020) + total_females(2020))

print((total_females(2020) / freq_sum(2020))*100)

print((total_males(2020) / freq_sum(2020))*100)
`````

`````python
def pc_females(year):
    """Provides the percent of females babies for a given year, """
    data_yr = parse_dataset(year)
    pop_total = data_yr['frequency'].sum()
    female_df = data_yr[data_yr['gender'] != 'M']
    ttl_females = female_df['frequency'].sum()
    pc_fem = round((ttl_females / pop_total) * 100,2)
    return pc_fem

pc_females(2020)
`````

`````python
def pc_males(year):
    """Provides the percent of males babies for a given year, """
    data_yr = parse_dataset(year)
    pop_total = data_yr['frequency'].sum()
    male_df = data_yr[data_yr['gender'] != 'F']
    ttl_males = male_df['frequency'].sum()
    pc_men = round((ttl_males / pop_total) * 100,2)
    return pc_men

pc_males(2020)
`````
