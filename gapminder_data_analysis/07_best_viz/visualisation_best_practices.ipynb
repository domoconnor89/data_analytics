### Importing the libraries

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

### Let's read the Data

```python
penguins = pd.read_csv('./data/penguins_year.csv')
penguins.head()
```

Notice that we have another column in this data set - `year` (which is not the part of the publicly available dataset that we have seen before but it's the real information). Penguins data were collected and made available by Dr. Kristen Gorman and the Palmer Station, Antarctica LTER, a member of the Long Term Ecological Research Network. Let's use this data to explore more visualisation options with seaborn and matplotlib!

### Remember the question from the seaborn encounter: 
### Q: Flipper Length for 3 Penguin Species ?

```python
sns.barplot(x ="species", y="flipper_length_mm", data=penguins);
```

#### Let's add a title with styling

```python
sns.barplot(x ="species", y="flipper_length_mm", data=penguins)
plt.title("Flipper Length for 3 Penguin Species", size=18, color="purple");
```

#### Let's play with the size and the color with using one of the color parameter

```python
plt.subplots(figsize=(8,6))
sns.barplot(x ="species", y="flipper_length_mm", data=penguins, color='c')
plt.title("Flipper Length for 3 Penguin Species", size=18, color="purple");
```

#### Using one of the color pallete(instead of one color) and adding edgecolor and line width

```python
#Examples:
#Purples', 'Purples_r','RdGy', 'RdGy_r', 'RdPu', 'RdPu_r', 'RdYlBu', 'RdYlBu_r', 'RdYlGn'
plt.subplots(figsize=(8,6))
sns.barplot(x ="species", y="flipper_length_mm", data=penguins, 
            palette='RdPu_r', 
#             edgecolor ='mediumpurple',
            lw =3)
plt.title("Flipper Length for 3 Penguin Species", size=18, color="purple");
```

#### Can we add more categories? and make it more readable

```python
plt.subplots(figsize=(7,5))
sns.barplot(x ="species", y="flipper_length_mm", hue='sex', data=penguins, 
            palette='RdPu_r', 
            edgecolor ='mediumpurple', 
            lw =3)
plt.title("Flipper Length for 3 Penguin Species", size=18, color="purple");
```

```python
# But first a reminder how are the bars calculated

adelie = penguins[penguins['species']== 'Adelie']
adelie.describe()

# default, it is mean
```

#### Let's add the values on the graph

```python
plt.subplots(figsize=(8,6))
ax = sns.barplot(x ="species", y="flipper_length_mm", hue='sex', errorbar=None, data=penguins, 
            palette='RdPu_r', 
            edgecolor='mediumpurple',
            lw=3)
for i in ax.containers:
    ax.bar_label(i,)
plt.title("Flipper Length for 3 Penguin Species", size=18, color="purple");
```

#### We need a bigger plot and also we need to put the legend outside

```python
plt.subplots(figsize=(10,8))
ax = sns.barplot(x ="species", y="flipper_length_mm", hue='sex', errorbar=None, data=penguins, 
            palette='RdPu_r', 
            edgecolor='mediumpurple',
            lw=3)
for i in ax.containers:
    ax.bar_label(i,)
plt.title("Flipper Length for 3 Penguin Species", size=18, color="purple")
plt.legend(bbox_to_anchor=(1.2,1));
```

#### Move the label in the middle of the bars

```python
plt.subplots(figsize=(8,6))
ax = sns.barplot(x ="species", y="flipper_length_mm", hue='sex', errorbar=None, data=penguins, 
            palette='RdPu_r', 
            edgecolor='mediumpurple',
            lw=3)
ax.bar_label(ax.containers[0], size=13,label_type='center', color ='pink') # we also changed the color and the location
ax.bar_label(ax.containers[1], size=13,label_type='center', color ='purple')
plt.title("Flipper Length for 3 Penguin Species", size=18, color="purple")
plt.legend(bbox_to_anchor=(1.2,1));
```

### Q: Does the trend between body mass and flipper length hold for both sexes ?

```python
# We used the hue parameter and set it to sex

sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', data=penguins, hue='species'
               ).set(title='Do bigger pinguins have longer flippers ?');

# Notice that we are passing the title in seaborn not with matplotlib
```

#### Figure size

```python
plt.subplots(figsize=(8,4))
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', data=penguins, hue='species'
               ).set(title='Do bigger penguins have longer flippers ?');
```

#### What else did we use? Let's add them also into the scatterplot: size, color?, and changing the place of the legend

### Let's customize the way the plot looks

```python
#Examples for palettes:
#icefire', 'icefire_r', 'inferno', 'inferno_r', 
#'jet', 'jet_r', 'magma', 'magma_r', 'mako', 'mako_r', 'nipy_spectral'

plt.subplots(figsize=(8,6))

sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', data=penguins, hue='species',
               palette='magma_r').set(title='Do bigger penguins have longer flippers ?');
plt.legend(loc='lower right'); #changing the place of the legend
```

#### Adding grid

```python
sns.set_style("darkgrid") # setting the style
plt.subplots(figsize=(8,6))
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', data=penguins, hue='species',
               palette='magma_r').set(title='Do bigger penguins have longer flippers ?');
plt.legend(loc='upper right', frameon=False); # removing the frame around the legend
```

#### Color cannot be enough to distinguish, also maybe for the colorblind people

```python
plt.subplots(figsize=(8,6))
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', data=penguins, hue='species',
               palette='magma_r', style='species').set(title='Do bigger penguins have longer flippers ?');
plt.legend(loc='upper right', frameon=False);
```

#### Let's make them bigger and play with the transparency

```python
plt.subplots(figsize=(8,6))
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', data=penguins, hue='species',
               palette='magma_r', s=200, style='species', alpha = 0.5).set(title='Do bigger pinguins have longer flippers ?');
plt.legend(loc='upper right', frameon=False);
```

#### We can also choose which markers we would like to have

```python
plt.subplots(figsize=(8,6)) 
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', data=penguins, hue='species',
               palette='magma_r', s=120, 
               style='species', alpha = 0.5, markers = ['D','o','v']).set(title='Do bigger pinguins have longer flippers ?');
plt.legend(loc='upper right', frameon=False);

# other possible markers:
# '.', ',', 'o', 'v', '^', '<', '>', '1', '2', '3', '4', '8', 's', 'p', '*', 'h', 'H', '+', 'x', 'D', 'd', '|', '_', 'P', 'X'
```

### Drawing the scatter plot for year 2007

```python
mass= df_2007['body_mass_g'] #.unique()

plt.subplots(figsize=(10,8))
ax = sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', data=df_2007, hue='species',
               palette='magma_r', 
               style='species', alpha = 0.5, markers = ['D','o', 'v'], size='body_mass_g', sizes=(10, 250)).set(title='Do bigger pinguins have longer flippers ?');
plt.legend(bbox_to_anchor=(1.2,1), frameon = False)
#plt.legend(bbox_to_anchor=(1.4,1), frameon = False)
plt.xticks(rotation=45);
```

#### Let's display the values on the graph

```python
mass= df_2007['body_mass_g']#.unique()

plt.subplots(figsize=(10,8))
ax = sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', data=df_2007, hue='species',
               palette='magma_r', 
               style='species', alpha = 0.5, markers = ['D','o','v'], size='body_mass_g', sizes=(10, 250)).set(title='Do bigger pinguins have longer flippers ?');
plt.legend(bbox_to_anchor=(1.2,1), frameon = False)
#plt.legend(bbox_to_anchor=(1.4,1), frameon = False)
plt.xticks(rotation=45);


for i in mass:
    x_cor = (df_2007.loc[df_2007['body_mass_g']== i, ['bill_length_mm']].values[0])
    y_cor = (df_2007.loc[df_2007['body_mass_g']== i, ['bill_depth_mm']].values[0])
    plt.text(s = i, x = x_cor, y = y_cor, fontsize =7, weight = 'bold');
    
#df.values returns a numpy array with the underlying data of the DataFrame, without any index or columns names.

#(the first column is index 0)
#df_2007.loc[df_2007['species']== i, ['body_mass_g']].values[0] BUNU DENE
```
