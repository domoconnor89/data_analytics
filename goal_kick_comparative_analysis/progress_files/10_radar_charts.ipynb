```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from mplsoccer import PyPizza, add_image, FontManager

from scipy import stats
import math
```

```python
def get_pizza_plot(team,season):
    df = pd.read_csv(f'../data/KPIs_tbl_all.csv')
    
    df = df[['team','season', 'total_GKs', 'short_GK_pc', 'shrt_scsfl_GKs',
             'short_avg_x_prog', 'short_shot_pc','short_goal_pc',
             'fllw_shrt_goals_pc','fllw_shrt_shots_pc',
             'short_agg_gains', 'avg_GK_dist', 'ttl_avg_chain_speed',
             'ttl_prgpass_pcnt','ttl_prgpass_avg_dist','long_agg_gains',
             'fllw_long_shots_pc','fllw_long_goals_pc',
             'long_goal_pc','long_shot_pc','long_avg_x_prog','lng_scsfl_GKs',
             'ttl_avg_num_passes']]
    
    params = list(df.columns)
    params = params[2:]
    
    df_ssn = df[df['season'] == f'{season}'].reset_index(drop=True)
    
    df_team = df_ssn.loc[df_ssn['team'] == f'{team}'].reset_index(drop=True)
    
    team_vals = list(df_team.loc[0])[2:]
    
    values = []
    
    for x in range (len(params)):
        values.append(math.floor(stats.percentileofscore(df_ssn[params[x]],team_vals[x])))
    
    team_colour = team.replace(' ', '').replace('&', '')
    get_team_colours(team_colour)
    c1 = get_team_colours(team_colour)[0]
    m1 = get_team_colours(team_colour)[1]
    c2 = get_team_colours(team_colour)[2]
    l1 = get_team_colours(team_colour)[3]
    c3 = get_team_colours(team_colour)[4]
    m2 = get_team_colours(team_colour)[5]
    c4 = get_team_colours(team_colour)[6]
    l2 = get_team_colours(team_colour)[7]
    
    
    font_normal = FontManager('https://raw.githubusercontent.com/google/fonts/main/apache/roboto/'
                              'Roboto%5Bwdth,wght%5D.ttf')
    font_italic = FontManager('https://raw.githubusercontent.com/google/fonts/main/apache/roboto/'
                              'Roboto-Italic%5Bwdth,wght%5D.ttf')
    font_bold = FontManager('https://raw.githubusercontent.com/google/fonts/main/apache/robotoslab/'
                            'RobotoSlab%5Bwght%5D.ttf')

    
    # instantiate PyPizza class
    baker = PyPizza(
        params=params,                  # list of parameters
        background_color= 'none',
        straight_line_color = c2,  # color for straight lines
        straight_line_lw=5,             # linewidth for straight lines
        last_circle_lw=9,               # linewidth of last circle
        last_circle_color= c3,
        other_circle_lw=3,              # linewidth for other circles
        other_circle_ls="-.",            # linestyle for other circles
        other_circle_color= l1,
    )

    # plot pizza
    fig, ax = baker.make_pizza(
        values,              # list of values
        figsize=(10, 10),      # adjust figsize according to your need
        param_location=110,  # where the parameters will be added
        kwargs_slices=dict(
            facecolor= c1, edgecolor = c2,
            zorder=2, linewidth=6
        ),                   # values to be used when plotting slices
        kwargs_params=dict(
            color="none", font = 'Avenir',fontsize=18,
        fontproperties=font_bold.prop, va="center"
        ),                   # values to be used when adding parameter
        kwargs_values=dict(
            color= l2,  font = 'Avenir',fontsize=18, fontweight='bold', zorder=3,
            bbox=dict(
                edgecolor= m2, facecolor= m1,
                boxstyle="round,pad=0.2", lw=3
            )
        )                    # values to be used when adding parameter-values
    )
    
    #plt.savefig(f'../pitch_maps/pizza_plots/1819/pizza_{team_colour}.png');
    plt.show();
```
