**4.** Extract features like month, day, etc. from the Order Date datetime column into their own columns.

```python
## check the order_date dtype

df_orders.dtypes
```

```python
## convert order_date dtype to 'datetime'

df_orders['order_date'] = pd.to_datetime(df_orders['order_date'], dayfirst = True)

df_orders.dtypes
```

```python
## Check the new output

df_orders.head()
```

```python
## adding a year column

df_orders['year'] = df_orders['order_date'].dt.year
df_orders
```

```python
## adding a month column

df_orders['month'] = df_orders['order_date'].dt.month
df_orders
```

```python
## adding a DoW column

df_orders['day'] = df_orders['order_date'].dt.day_name()
df_orders
```

```python
## Setting bins for the annual quarters

bin_labels = ['Q1', 'Q2', 'Q3', 'Q4']
```

```python
## Setting the monthly limits for each quarters

bin_boundaries = [0,3,6,9,12]
```

```python
df_orders['quarter'] = pd.cut(df_orders['month'], bins = bin_boundaries, labels = bin_labels)
```

**5.** Plot small sections of the order date over time (1 week, 1 month, etc.)

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
```

```python
jan_19 = df_orders[(df_orders['year']==2019)&(df_orders['month']==1)]
```

```python
jan_19 = jan_19.groupby('order_date')['sales'].sum().reset_index()
```

```python
sns.lineplot(x='order_date', y='sales', data=jan_19)

plt.xticks(rotation=70);
```

**6.** Plot the number of orders per weekday

```python
DoW = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']

sns.lineplot(x='day', y='sales', data=df_orders)

plt.xticks(rotation=90);
```

```python
dow_sales = pd.DataFrame(df_orders.groupby('day')['sales'].sum()).reset_index()
```

```python
sns.lineplot(x='day', y='sales', data=dow_sales)

plt.xticks(rotation=90);
```

```python
dow_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

dow_sales['day'] = pd.Categorical(dow_sales['day'], categories=dow_order, ordered=True)
dow_sales = dow_sales.sort_values(by='day')
dow_sales
```

```python
sns.lineplot(x='day', y='sales', data=dow_sales)

plt.xticks(rotation=90);
```
